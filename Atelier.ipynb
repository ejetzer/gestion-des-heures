```python
activités_df = pd.read_excel(str(activités), sheet_name=['Janvier', 'Février', 'Mars', 'Avril', 'Mai', 'Janvier 2021'], header=4, usecols='A:H')

tempo = pd.read_excel(str(activités), sheet_name=['Résumé 16 février', 'Résumé 16 février 2', 'Résumé 16 février 3'], header=0, usecols='B:E')
for clé, df in tempo.items():
    df = df.rename(columns={'Description': 'Description des travaux effectués', 'Temps (min)': 'Nbr d\'heures '})
    df.loc[:, 'Nbr d\'heures '] = df['Nbr d\'heures '] / 60.0
    tempo[clé] = df
activités_df.update(tempo)
    
activités_df.keys()
```

```python
heures_atelier = aglo_édité[aglo_édité.Atelier == True].loc[:, ['Payeur', 'Date', 'Description des travaux effectués', 'Nbr d\'heures ']]
```

```python
atelier = df.loc[df.Atelier.astype(bool), ['Payeur',
                                           'Date',
                                           'Description des travaux effectués',
                                           'Nbr d\'heures ']]
minutes = df.loc[df.Atelier.astype(bool), 'Temps (min)'] / 60.0
atelier['Nbr d\'heures '] = atelier['Nbr d\'heures '].str\
                                                     .replace(',', '.')\
                                                     .astype(float)\
                                                     .fillna(minutes)\
                                                     .multiply(1.5)
atelier.head()
```

```python
heures_atelier.loc[:, 'Nbr d\'heures '] = heures_atelier.loc[:, 'Nbr d\'heures '] * 1.5
```

```python
heures_atelier = heures_atelier.sort_values(by=['Payeur', 'Date'])
```
